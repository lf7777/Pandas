## 1.函数实现

```python
df = pd.DataFrame(np.random.randint(1,100,60).reshape(10,6))
```

```python
df.rename(columns={3:'年龄'},inplace=True)
```

```python
df2 = np.random.choice(['后卫','门将','前锋','中场','摇摆'],size=(10),replace=True)
```

```python
df2
```

```python
df['位置']=df2
```

```python
pos = df.groupby('位置').sum()
```

```python
pos1 = pos.index
pos1
```

```python
for item in pos1:
    df['位置_'+item] = (df['位置'] == item).astype('int')
```

```python
df
```

```python
# 前锋中场分一组, 后卫门将分一组
```

```python
df = pd.DataFrame(np.random.randint(1,100,60).reshape(10,6))
```

```python
df.rename(columns={3:'年龄'},inplace=True)
```

```python
df2 = np.random.choice(['后卫','门将','前锋','中场'],size=(10),replace=True)
```

```python
df['位置']=df2
```

```python
df
```

```python
pos =[['位置','前锋','中场'],['位置','后卫','门将']]
```

```python
for item in pos:
    df['{}_{}&{}'.format(item[0],item[1],item[2])] = (df['位置'].isin(item)).astype('int')
```

```python
df
```

## 2.pd.get_dummies() 实现独热编码

```python
df = pd.DataFrame({'key':['b','b','a','c','a','d'],'data1':range(6)})
df
```

```python
dr = pd.get_dummies(df['key'],prefix='key') # prefix 以何作为 独热编码列名的 前缀
dr
```

```python
df_with_dummy = df.join(dr)
```

```python
df_with_dummy
```
