```python
#准备数据
df = pd.DataFrame(np.random.rand(10,4))
```

```python
df
```

```python
x = df[:2]
y = df[3:5]
z = df[7:]
```

```python
pd.concat([x,y,z])
```

```python
x = df[[0,2]]
y = df[[1,3]]
```

```python
pd.concat([x,y],axis=1)
```

```python
# merge
pd.merge(df,y)
```
